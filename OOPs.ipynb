```python
# Q1
class Emp:
  EmpId = None
  EmpName = None
  Points = 0
  Group = None 
  Average_Points = None
  silver = []
  gold = []
  platinum = []
  diamond = []
  def __init__(self, id = 3, name = 'Mahek'):
    self.EmpId = id
    self.EmpName = name
  def addPoints(self, p1):
    self.Points += p1
  def removePoints(self, p2):
    self.Points -= p2
    if (self.Points < 0):
      self.Points = 0
  def computeGroup(self):
    if (self.Points <= 100):
      self.Group = 'Silver'
      self.silver.append(self.EmpName)
    elif (self.Points>100 and self.Points<=500):
      self.Group = 'Gold'
      self.gold.append(self.EmpName)
    elif (self.Points>500 and self.Points<=1000):
      self.Group = 'Platinum'
      self.platinum.append(self.EmpName)
    elif (self.Points>1000):
      self.Group = 'Diamond'
      self.diamond.append(self.EmpName)
  def list_of_emp(self, g):
    if (g == 'Silver'):
      print(self.silver)
    elif (g == 'Gold'):
      print(self.gold)
    elif (g == 'Platinum'):
      print(self.platinum)
    elif (g == 'Diamond'):
      print(self.diamond)
    else:
      print("Wrong group entered!")
  def display(self):
    print("Employee ID: ", self.EmpId)
    print("Employee Name: ", self.EmpName)
    print("Employee Points: ", self.Points)
    print("Employee Group: ", self.Group)
    # print("Average Points: ", self.Average_Points)
    print("")
```

```python
e1 = Emp()
e2 = Emp(6, 'Nikita')
e3 = Emp(9, 'Muskan')
e4 = Emp(12, 'Shruti')
e1.addPoints(100)
e2.addPoints(300)
e3.addPoints(600)
e4.addPoints(1200)
e1.removePoints(10)
e2.removePoints(310)
e3.removePoints(20)
e4.removePoints(35)
e1.computeGroup()
e2.computeGroup()
e3.computeGroup()
e4.computeGroup()
e1.display()
e2.display()
e3.display()
e4.display()
e1.list_of_emp('Silver')
```

```python
# Q2
class Property:
    def __init__(self):
        self.square_footage = int(input("Enter square footage: "))
        self.no_of_bedrooms = int(input("Enter no. of bedrooms: "))
        self.no_of_bathrooms = int(input("Enter no. of bathrooms: "))
    def display(self):
        print("Property Details: ")
        print("Square footage: ", self.square_footage)
        print("No. of bedrooms: ", self.no_of_bedrooms)
        print("No. of bathrooms: ", self.no_of_bathrooms)
        
class House(Property):
    def __init__(self):
        super().__init__()
        self.no_of_stories = int(input("Enter no. of stories: "))
        self.garage = input("Enter type of garage (Attached/Detached/None): ")
        self.yard = input("Is the yard fenced? (Yes/No): ")
    def display(self):
        super().display()
        print("House Details: ")
        print("No. of stories: ", self.no_of_stories)
        print("Garage: ", self.garage)
        print("Yard: ", self.yard)
        
class Apartment(Property):
    def __init__(self):
        super().__init__()
        self.balcony = input("Is there a balcony? (Yes/No): ")
        self.laundry = input("Enter type of Laundry (En-suite/Coin/Off-site): ")
    def display(self):
        super().display()
        print("Apartment Details: ")
        print("Balcony: ", self.balcony)
        print("Laundry: ", self.laundry)
        
class HouseRental(House):
    def __init__(self):
        super().__init__()
        self.rent_per_month = int(input("Enter rent per month: "))
        self.utilities = input("Are utilities included? (Yes/No): ")
        if (self.utilities == "No"):
          self.estimated_utilitites = input("What are the estimated utilities?: ")
    def display(self):
        super().display()
        print("Rental Details: ")
        print("Rent per month: ", self.rent_per_month)
        print("Utilities included: ", self.utilities)
        if (self.utilities == "No"):
          print("Estimated utilities: ", self.estimated_utilities)
        
class HousePurchase(House):
    def __init__(self):
        super().__init__()
        self.price = int(input("Enter price: "))
        self.estimated_annual_property_tax = int(input("Enter estimated annual property tax: "))
    def display(self):
        super().display()
        print("Purchase Details: ")
        print("Price: ", self.price)
        print("Estimated annual property tax: ", self.estimated_annual_property_tax)
        
class ApartmentRental(Apartment):
    def __init__(self):
        super().__init__()
        self.rent_per_month = int(input("Enter rent per month: "))
        self.utilities = input("Are utilities included? (Yes/No): ")
        if (self.utilities == "No"):
          self.estimated_utilitites = input("What are the estimated utilities?: ")
    def display(self):
        super().display()
        print("Rental Details: ")
        print("Rent per month: ", self.rent_per_month)
        print("Utilities included: ", self.utilities)
        if (self.utilities == "No"):
          print("Estimated utilities: ", self.estimated_utilities)
        
class ApartmentPurchase(Apartment):
    def __init__(self):
        super().__init__()
        self.price = int(input("Enter price: "))
        self.estimated_annual_property_tax = int(input("Enter estimated annual property tax: "))
    def display(self):
        super().display()
        print("Purchase Details: ")
        print("Price: ", self.price)
        print("Estimated annual property tax: ", self.estimated_annual_property_tax)
        
class Agent:
    list1 = []
    def addProperty(self):
        str = input("Enter type of property (House/Apartment): ")
        if (str == "House"):
            s = input("Enter payment type (Rental/Purchase): ")
            if (s == "Rental"):
              obj = HouseRental()
              Agent.list1.append(obj)
            elif (s == "Purchase"):
              obj = HousePurchase()
              Agent.list1.append(obj)
        elif (str == "Apartment"):
            s = input("Enter payment type (Rental/Purchase): ")
            if (s == "Rental"):
              obj = ApartmentRental()
              Agent.list1.append(obj)
            elif (s == "Purchase"):
              obj = ApartmentPurchase()
              Agent.list1.append(obj)
                
    def displayProperty(self):
        for i in Agent.list1:
            print()
            i.display()
            print()
        
obj1 = Agent()
obj1.addProperty()
obj1.addProperty()
obj1.displayProperty()
```

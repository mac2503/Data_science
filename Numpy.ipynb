```python
# Q1
import numpy as np
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data = np.genfromtxt(url, delimiter=',')
data
```

```python
len(data)
```

```python
data = np.delete(data, 4, axis=1)
data
```

```python
col = []
for i in range(0,4):
  t = int(input())
  col.append(t)
col
```

```python
def Sort(sub_list):
    l = len(sub_list)
    for i in range(0, l):
        for j in range(0, l-i-1):
            if (sub_list[j][1] > sub_list[j + 1][1]):
                temp = sub_list[j]
                sub_list[j] = sub_list[j + 1]
                sub_list[j + 1] = temp
    return sub_list
```

```python
diff = []
# Count-> To store index
count = 0 
for i in data:
  list1 = i
  n_list = []
  n_list.append(count)
  # To store distance
  n_list.append(np.linalg.norm(list1 - col))
  diff.append(n_list)
  count = count+1
print(diff)
```

```python
Sort(diff)
```

```python
indexes = []
count = 0
for i in diff:
  if(count>=5):
    break
  indexes.append(i[0])
  count = count+1
indexes
```

```python
from math import sqrt
# Create function to calculate Manhattan distance 
# The Manhattan Distance between two points (X1, Y1) and (X2, Y2) is given by |X1 – X2| + |Y1 – Y2|.
def manhattan(a, b):
    return sum(abs(val1-val2) for val1, val2 in zip(a,b))
```

```python
# Q2
manhattan_diff = []
count = 0
for i in data:
  n_list = []
  list1 = i
  n_list.append(count)
  md = manhattan(list1, col)
  n_list.append(md)
  manhattan_diff.append(n_list)
  count = count+1
manhattan_diff
```

```python
indexes = []
for i in range(0, 15):
  indexes.append(i)
indexes
```

```python
def sortLists(list1, indexes):
    for i in range(0, 15):
        for j in range(0, 15-i-1):
            if (list1[j] < list1[j + 1]):
                temp = list1[j]
                list1[j] = list1[j + 1]
                list1[j + 1] = temp
                t = indexes[j]
                indexes[j] = indexes[j+1]
                indexes[j+1] = t
    return list1, indexes
```

```python
new_list, new_index = sortLists(row_mean, indexes)
new_list
```

```python
new_index
```

```python
print("Best case: ")
print("Roll no.: ", mod_df.iloc[new_index[0]][0])
print("Mean marks: ", new_list[0])
```

```python
med = new_list.median()
med
```

```python
ind = 0
print("Average case: ")
for i in range(0, 16):
  if (new_list[i]==med):
    ind = i
    break
print("Roll no.: ", mod_df.iloc[new_index[ind]][0])
print("Mean marks: ", med)
```

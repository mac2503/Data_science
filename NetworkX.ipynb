```python
# Q1
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 7, 9])
nx.draw(G, with_labels=True, font_color='g', node_color='lightgrey', node_size=600)
```

```python
# Q2
G.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 7), (1, 9), (2, 4), (2, 9)])
nx.draw(G, with_labels = True, font_color='g', node_color='lightgrey', node_size=600)
```

```python
# Q3
G.remove_node(3)
G.remove_edge(1, 2)
nx.draw(G, with_labels = True, font_color='g', node_color='lightgrey', node_size=600)
```

```python
# Q4
H = nx.Graph()
H.add_nodes_from(["City A", "City B", "City C", "City D"])
H.add_edge("City A", "City B", weight = 10)
H.add_edge("City A", "City C", weight = 15)
H.add_edge("City A", "City D", weight = 20)
H.add_edge("City B", "City C", weight = 35)
H.add_edge("City B", "City D", weight = 25)
H.add_edge("City C", "City D", weight = 30)

H.nodes["City A"]['pos'] = (2, 8)
H.nodes["City B"]['pos'] = (0, 0)
H.nodes["City C"]['pos'] = (4, 0)
H.nodes["City D"]['pos'] = (2, 4)

node_pos = nx.get_node_attributes(H, 'pos')
arc_weight = nx.get_edge_attributes(H, 'weight')

nx.draw(H, node_pos, with_labels=True, edgecolors='black', font_color='orange', node_color= 'white', node_size=2100)
nx.draw_networkx_edges(H, node_pos, edge_color='k')
nx.draw_networkx_edge_labels(H, node_pos, edge_labels=arc_weight)
```

```python
# Q5
G = nx.DiGraph()
G.add_nodes_from(["A", "B", "C", "D"])
G.add_edge("A", "B", weight = 4)
G.add_edge("A", "C", weight = 12)
G.add_edge("A", "D", weight = 7)
G.add_edge("B", "A", weight = 5)
G.add_edge("B", "D", weight = 18)
G.add_edge("C", "A", weight = 11)
G.add_edge("C", "D", weight = 6)
G.add_edge("D", "A", weight = 10)
G.add_edge("D", "B", weight = 2)
G.add_edge("D", "C", weight = 3)

G.nodes["A"]['pos'] = (1, 8)
G.nodes["B"]['pos'] = (0, 0)
G.nodes["C"]['pos'] = (5, 8)
G.nodes["D"]['pos'] = (4, 0)

pos = nx.get_node_attributes(G, 'pos')

nx.draw(G, pos, with_labels=True,  edgecolors='black', font_color='orange', node_color='white', node_size=600, connectionstyle='arc3, rad = 0.1')
edge_labels = dict([((u, v), f'{d["weight"]}\n\n{G.edges[(v,u)]["weight"]}') for u, v, d in G.edges(data=True) if pos[u][0] > pos[v][0]])

nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='orange')
plt.show()
```
